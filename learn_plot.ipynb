```python
import numpy as np
import matplotlib.pyplot as plt

fig_name_suffix_ = 'learn_plot_'
```

```python
x_list = [1,2,3,4,5]
y_list = [1,1,2,3,5]


plt.plot(x_list, y_list, linewidth=2.0)
plt.xlim([np.min(x_list)*1.05, np.max(x_list)*1.05])
plt.ylim([np.min(y_list), np.max(y_list)])

plt.title("Title")
plt.ylabel(r'Y', fontsize=18)
plt.xlabel(r'X', fontsize=18)
plt.savefig(f"figure/{fig_name_suffix_}list-list.png")
plt.show()
```
